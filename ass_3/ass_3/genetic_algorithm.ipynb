```python
RUL = pd.read_csv('data/RUL_consultancy_predictions_A3-2.csv', sep = ";")
RUL
```

```python
def select_nr_engines(RUL: pd.DataFrame, max_rul: int): 
    RUL_filtered = RUL[RUL["RUL"] <= max_rul].reset_index(drop = True)
    return RUL_filtered, len(RUL_filtered)


selected_engines, nr_engines = select_nr_engines(RUL, 29)
selected_engines
selected_engines["RUL"].tolist()
```

```python
def initialize_population(nr_engines: int, population_size: int, RUL: pd.DataFrame):
    random_teams = np.random.choice(range(1, G+1), size = nr_engines)    
    random_start_date = np.random.choice(range(1, T), size = nr_engines)
    merged = list(zip(random_teams, random_start_date))
    
    id_engine_list = selected_engines['id'].tolist()
    
    initital_schedule_engine = {key: value for key, value in zip(id_engine_list, merged)}   
    
    return [initital_schedule_engine for _ in range(population_size)]
    
initialize_population(nr_engines, population_size, RUL)
```

```python
binary_values = []

# Loop van 0 tot 29 (30 waarden)
for i in range(30):
    # Converteer de huidige waarde van i naar een binaire string met behulp van de bin() functie
    # en verwijder het '0b' prefix
    binary_value = bin(i)[2:]

    # Zorg ervoor dat elke binaire waarde 5 bits lang is door nullen aan de linkerkant toe te voegen indien nodig
    binary_value = binary_value.zfill(5)

    # Voeg de gegenereerde binaire waarde toe aan de lijst
    binary_values.append(binary_value)

# Print de lijst met binaire waarden
print(binary_values)
```
